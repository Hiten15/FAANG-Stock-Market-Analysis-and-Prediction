# Time Resampling

```python
#see this article - https://towardsdatascience.com/resample-function-of-pandas-79b17ec82a78#:~:text=Resampling%20is%20used%20in%20time,DatetimeIndex%2C%20PeriodIndex%2C%20or%20TimedeltaIndex.
```

```python
df_tesla.head()
```

```python
#rule A means year end frequency
df_tesla.resample(rule='A').min()    #mtlb har sall ke aakhri mahine ki aakhri date pe min open,high,low vgera
```

```python
df_tesla.resample(rule='A').max()
```

```python
df_tesla.resample(rule='A').max()['Open'].plot()
```

```python
#quaterly start frequency
df_tesla.resample(rule='QS').max()
```

```python
df_tesla.resample(rule='QS').max()['High'].plot()
```

```python
#Buisness end frequency
df_tesla.resample(rule='BA').max()
```

```python
#Buisness quarter start freq
df_tesla.resample(rule='BQS').max()
```

```python
df_tesla['Open'].resample(rule='A').mean().plot(kind='bar')
```

```python
df_tesla['Open'].resample(rule='M').max()  #monthly max
```

```python
df_tesla['Open'].resample(rule='M').max().plot(kind='bar',figsize=(25,10))
```

```python
#Rolling means creating a rolling window with a specified size and perform calculations on the data in this window which, 
#of course, rolls through the data.
#see article - https://towardsdatascience.com/time-series-analysis-resampling-shifting-and-rolling-f5664ddef77e
```

```python
df_tesla['High'].head(20)
```

```python
df_tesla['High'].rolling(11).max().head(20)
```

```python
df_tesla['Open:30 days rolling']=df_tesla['Open'].rolling(30).mean()
```

```python
df_tesla.head(31)
```

```python
df_tesla[['Open','Open:30 days rolling']].plot(figsize=(12,5))   #rolling smoothens the curve
```
